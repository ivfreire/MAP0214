```python
import numpy as np
import matplotlib.pyplot as plt
```

## MÃ©todo de Euler

```python
def g(t, y, z): return 12 * t**2 + 4 * t**3 - t**4 + y - z
```

```python
h = 0.01
y = 0
z = 0
t = 0
N = int(6 / h)
```

```python
axis = np.zeros((3, N))
```

```python
for i in range(N):
    y += h * z
    z += h * g(t, y, z)
    
    axis[0][i] = t
    axis[1][i] = y
    axis[2][i] = z
    
    t += h
```

```python
x = np.arange(0.5, 6, 0.5)
```

```python
plt.plot(axis[0], axis[1])
plt.scatter(x, x**4, color='red')
plt.title('GrÃ¡fico de y')
plt.ylabel('y')
plt.xlabel('t')
plt.grid(True)
plt.show()
```

```python
plt.plot(axis[0], axis[2])
plt.scatter(x, 4 * x**3, color='red')
plt.title('GrÃ¡fico de z')
plt.ylabel('z')
plt.xlabel('t')
plt.grid(True)
plt.show()
```

```python
for i in range(N):
    print('{}\t{}\t{}'.format(axis[0][i], axis[1][i], axis[2][i]))
```

## MÃ©todo de Runge-Kutta

```python
def g(t, y, z): return 12 * t**2 + 4 * t**3 - t**4 + y - z
```

```python
h = 0.01
y = 0
z = 0
t = 0
N = int(6 / h)
```

```python
axis = np.zeros((3, N))
```

```python
for i in range(N):
    k1y = h * z
    k1z = h * g(t, y, z)
    k2y = h * (z + k1z / 2)
    k2z = h * g(t + h / 2, y + k1y / 2, z + k1z / 2)
    k3y = h * (z + k2z / 2)
    k3z = h * g(t + h / 2, y + k2y / 2, z + k2z / 2)
    k4y = h * (z + k3z)
    k4z = h * g(t + h, y + k3y, z + k3z)
    
    y += (k1y + 2 * k2y + 2 * k3y + k4y) / 6
    z += (k1z + 2 * k2z + 2 * k3z + k4z) / 6
    
    axis[0][i] = t
    axis[1][i] = y
    axis[2][i] = z
    
    t += h
```

```python
x = np.arange(0.5, 6, 0.5)
```

```python
plt.plot(axis[0], axis[1])
plt.scatter(x, x**4, color='red')
plt.title('GrÃ¡fico de y')
plt.ylabel('y')
plt.xlabel('t')
plt.grid(True)
plt.show()
```

```python
plt.plot(axis[0], axis[2])
plt.scatter(x, 4 * x**3, color='red')
plt.title('GrÃ¡fico de z')
plt.ylabel('z')
plt.xlabel('t')
plt.grid(True)
plt.show()
```

```python
for i in range(N):
    print('{}\t{}\t{}'.format(axis[0][i], axis[1][i], axis[2][i]))
```
