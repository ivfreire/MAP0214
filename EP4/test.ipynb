```python
import numpy as np
import matplotlib.pyplot as plt
```

## Método de Euler

```python
def f(t, y):
    if t != 0: return np.exp(t)
    else: return 1
```

```python
h = 0.01
t = 0
y = 1
N = 100
```

```python
axis = np.zeros((2, N))
```

```python
for i in range(N):
    y += h * f(t, y)
    
    axis[0][i] = t
    axis[1][i] = y
    
    t += h
```

```python
x = np.arange(0.1, 1, 0.1)
```

```python
plt.plot(axis[0], axis[1])
plt.scatter(x, np.exp(x), color='red')
plt.show()
```

## Método de Runge-Kutta

```python
def f(t, y):
    if t != 0: return np.exp(t)
    else: return 1
```

```python
h = 0.01
t = 0
y = 1
N = 100
```

```python
axis = np.zeros((2, N))
```

```python
for i in range(N):
    k1 = h * f(t, y)
    k2 = h * f(t + h / 2, y + k1 / 2)
    y += (k1 + k2) / 2 # (k1 + k2 + k3 + k4) / 6
    
    axis[0][i], axis[1][i] = t, y
    
    t += h
```

```python
x = np.arange(0.1, 1, 0.1)
```

```python
plt.plot(axis[0], axis[1])
plt.scatter(x, np.exp(x), color='red')
plt.show()
```
