```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calculate(t, x, v, h, N, n=0):
    if n == 0: n = N
    
    axis = np.zeros((3, n))
    for i in range(N):
        k1x = h * v
        k1v = h * g(t, x, v)
        k2x = h * (v + k1v / 2)
        k2v = h * g(t + h / 2, x + k1x / 2, v + k1v / 2)
        k3x = h * (v + k2v / 2)
        k3v = h * g(t + h / 2, x + k2x / 2, v + k2v / 2)
        k4x = h * (v + k3v)
        k4v = h * g(t + h, x + k3x, v + k3v)

        x += (k1x + 2 * k2x + 2 * k3x + k4x) / 6
        v += (k1v + 2 * k2v + 2 * k3v + k4v) / 6

        if i >= N - n:
            axis[0][i - (N - n)] = t
            axis[1][i - (N - n)] = x
            axis[2][i - (N - n)] = v

        t += h
    return axis
```

## A

```python
def g(t, x, v): return (x - x**3) / 2
```

```python
axis0 = calculate(0, 1, -0.1, 0.01, 700)
```

```python
plt.plot(axis0[1], axis0[2])
plt.title('Espaço de Fase (v=-0.1)')
plt.xlabel('x(t)')
plt.ylabel('v(t)')
plt.grid()
plt.savefig('ep4_II_1a_0.png')
plt.show()
```
